```python
target_df['success'] = target_df.apply(lambda row: 'High' if row['possession_team_name'] == row['next_possession_2'] else 'Low', axis = 1)
```

```python
for idx, row in target_df.iterrows():
    receive_frame = row['receive_frame']
    pair_id = row['pair_id']
    frames = [f for f in data if f['frame'] == receive_frame+4 or f['frame']== receive_frame+5]
    frames_df = pd.DataFrame(frames)
    expanded_df = pd.concat(frames_df.apply(explode_data, axis=1).to_list(), ignore_index=True)
    ball_df = expanded_df[expanded_df['player_id'] == 0]
    pair_df = expanded_df[expanded_df['player_id'] == pair_id]    

    positions_A = ball_df[['x', 'y']].to_numpy()
    positions_B = pair_df[['x', 'y']].to_numpy()
    delta_t = 0.1
    velocities_A = (positions_A[1] - positions_A[0]) / delta_t
    velocities_B = (positions_B[1] - positions_B[0]) / delta_t    
    
    xy_1 = positions_A[1]
    xy_2 = positions_B[1]
    distance = lambda t: np.linalg.norm((xy_1 + velocities_A * t) - (xy_2 + velocities_B * t))
    result = minimize(distance, 0, bounds=[(0, None)])
    MPD = result.fun
    target_df.at[idx,'MPD'] = MPD
```
