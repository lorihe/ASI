```python
team_scores = data.groupby('possession_team_name')['score'].sum()
team_total = data['possession_team_name'].value_counts()
score_percentage = (team_scores / team_total) * 100
```

```python
# Create the first bar chart (team counts)
trace1 = go.Bar(
    x=team_total.values,
    y=team_total.index,
    orientation='h',
    name='1v1 Counts',
    marker=dict(color='steelblue')
)

# Create the second bar chart (team averages) below the first
trace2 = go.Bar(
    x=team_scores.values,
    y=team_scores.index,
    orientation='h',
    name='1v1 Success',
    marker=dict(color='tan')
)

# Create the figure
fig = go.Figure()
fig.add_trace(trace2)
fig.add_trace(trace1)

for i, value in enumerate(team_total.values):
    fig.add_annotation(
        x=value,
        y=team_total.index[i],
        text=f"{value}",
        showarrow=False,
        font=dict(size=12),
        xanchor="left",
        yanchor="bottom"
    )

# Add annotations for score percentages
for i, value in enumerate(score_percentage.values):
    fig.add_annotation(
        x=team_scores[i],
        y=score_percentage.index[i],
        text=f"{value:.1f}%",
        showarrow=False,
        font=dict(size=12),
        xanchor="left",
        yanchor="top"
    )

# Update layout to customize the appearance
fig.update_layout(
    title="Horizontal Bar Charts: Team Counts and Score Percentages",
    xaxis=dict(title="Values"),
    yaxis=dict(title="Teams"),
    barmode='group',  # Ensures the bars are clearly separated
    height=700,
    plot_bgcolor='white',  # Set background to white
)

# Adjust spacing between the bars
fig.update_yaxes(categoryorder='total ascending')

# Show the chart
fig.show()
```
