```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def w_exponential(t, lambd):
    return np.exp(-lambd * t)
def w_min_fraction(t, N):
    return np.minimum(1, N / t)
```

```python
lambd = 0.5 
N = 2  # Threshold value
t_values = np.linspace(1, 10, 10)
w_min_exp_values = np.minimum(1, w_exponential(t_values, lambd))
w_min_fraction_values = w_min_fraction(t_values, N)

# Plot comparison
plt.figure(figsize=(8,5))
plt.plot(t_values, w_min_exp_values, label='$w(t) = \\min(1, e^{-\\lambda t})$', color='blue')
plt.plot(t_values, w_min_fraction_values, label='$w(t) = \\min(1, \\frac{N}{t})$', color='purple')
plt.axhline(1, color='orange', linestyle='--', label='$w(t)=1$')
plt.title('Comparison of $\\min(1, e^{-\\lambda t})$ and $\\min(1, \\frac{N}{t})$')
plt.xlabel('$t$')
plt.ylabel('$w(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the piecewise function
def w_piecewise(t, lambd, threshold):
    return np.piecewise(
        t,
        [t <= threshold, t > threshold],
        [1, lambda t: np.exp(-lambd * t)]
    )

# Parameters
threshold = 2  # t <= 2
lambd = 0.5    # Decay rate for t > 2
t_values = np.linspace(0, 10, 501)  # Time values

# Compute w(t)
w_values = w_piecewise(t_values, lambd, threshold)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(t_values, w_values, label=f'$w(t)$ with $\\lambda={lambd}$, threshold={threshold}')
plt.axvline(threshold, color='red', linestyle='--', label=f'Threshold $t={threshold}$')
plt.title('Piecewise $w(t)$ Function')
plt.xlabel('$t$')
plt.ylabel('$w(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Adjust parameters
w_target = 0.5  # Desired w(t) value at t=3
t_target = 3    # The target time
lambd = -np.log(w_target) / t_target  # Calculate lambda based on target

# Compute w(t)
w_values = w_piecewise(t_values, lambd, threshold)

# Print lambda and w(t) at t=3 and t=5
w_at_3 = np.exp(-lambd * 3)
w_at_5 = np.exp(-lambd * 5)
print(f"Lambda: {lambd:.4f}")
print(f"w(t=3): {w_at_3:.4f}")
print(f"w(t=5): {w_at_5:.4f}")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(t_values, w_values, label=f'$w(t)$ with $\\lambda={lambd:.3f}$, threshold={threshold}')
plt.axvline(threshold, color='red', linestyle='--', label=f'Threshold $t={threshold}$')
plt.title('Adjusted Piecewise $w(t)$ Function')
plt.xlabel('$t$')
plt.ylabel('$w(t)$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Parameters
t_values = np.arange(1, 10 + 1)
time_gaps = [1, 20, 50, 100]  # Example time gaps
lambda_decay = 0.6
mu_decay = 0.02

# Plot
plt.figure(figsize=(10, 6))
for time_gap in time_gaps:
    weights = [min(1, np.exp(-lambda_decay * t)) * np.exp(-mu_decay * time_gap) for t in t_values]
    plt.plot(t_values, weights, label=f'time_gap={time_gap}')

plt.title('Weight Function with Time Gap')
plt.xlabel('Events Away from Shot (t)')
plt.ylabel('Weight')
plt.legend()
plt.grid()
plt.show()
```
