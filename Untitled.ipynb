```python
def plot_frame(i):
    frames = [f for f in data if f['frame']>= i-10 and f['frame']<=i+10]
    for frame in frames:
        ball = frame['ball_data']
        players_detected = [(p['x'], p['y']) for p in frame['player_data'] if p['is_detected']]
        players_undetected = [(p['x'], p['y']) for p in frame['player_data'] if not p['is_detected']]
        
        # Create pitch
        pitch = Pitch(pitch_type='skillcorner', pitch_length=105, pitch_width=68, linewidth = 1)
        fig, ax = pitch.draw(figsize=(6, 4))
        
        # Plot players
        pitch.scatter(*zip(*players_detected), ax=ax, c='green', label='Detected Players', s=10)
        pitch.scatter(*zip(*players_undetected), ax=ax, c='red', label='Undetected Players', s=10)
        
        # Plot ball
        if ball['is_detected']:
            pitch.scatter(ball['x'], ball['y'], ax=ax, c='orange', label='Ball', s=10, edgecolors='black')
        
        # Add legend
        ax.legend(loc='upper left', fontsize=10)
        print(frame['frame'])
        # Show the plot
        plt.show()    
```

```python
plot_frame(104)
```
