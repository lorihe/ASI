```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def w_exponential(t, lambd):
    return np.exp(-lambd * (t - 1))
def w_min_fraction(t, N):
    return np.minimum(1, N / t)
```

```python
plt.figure(figsize=(5,5))
t_values = np.linspace(1, 5, 5)

w_min_exp_values = np.minimum(1, w_exponential(t_values, 2))
plt.plot(t_values, w_min_exp_values, label='$w(t) = \\min(1, e^{-\\ 2 t})$', color='blue')
w_min_exp_values = np.minimum(1, w_exponential(t_values, 0.8))
plt.plot(t_values, w_min_exp_values, label='$w(t) = \\min(1, e^{-\\ 0.8 t})$', color='seagreen')
w_min_exp_values = np.minimum(1, w_exponential(t_values, 0.5))
plt.plot(t_values, w_min_exp_values, label='$w(t) = \\min(1, e^{-\\ 0.5 t})$', color='orange')

plt.xlabel('t')
plt.ylabel('w(t)')
plt.grid(True)
plt.legend()
plt.xticks(np.arange(1, 6, 1)) 
plt.show()
```

```python
# Parameters
t_values = np.arange(1, 10 + 1)
time_gaps = [1, 20, 50, 100]  # Example time gaps
lambda_decay = 0.6
mu_decay = 0.02

# Plot
plt.figure(figsize=(10, 6))
for time_gap in time_gaps:
    weights = [min(1, np.exp(-lambda_decay * t)) * np.exp(-mu_decay * time_gap) for t in t_values]
    plt.plot(t_values, weights, label=f'time_gap={time_gap}')

plt.title('Weight Function with Time Gap')
plt.xlabel('Events Away from Shot (t)')
plt.ylabel('Weight')
plt.legend()
plt.grid()
plt.show()
```
