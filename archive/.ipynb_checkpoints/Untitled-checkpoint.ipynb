```python
from mplsoccer.pitch import Pitch
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import jsonlines
```

```python
file_path = 'data/FA/tracking/1541631.jsonl'
```

```python
def load_jsonl_file(file_path):
    with jsonlines.open(file_path, 'r') as file:
        data = [line for line in file if line['timestamp'] is not None]
    return data
```

```python
data = load_jsonl_file(file_path)
```

```python
len(data)
```

```python
data_0 = data[0]
```

```python
ball = data_0['ball_data']
players_detected = [(p['x'], p['y']) for p in data_0['player_data'] if p['is_detected']]
players_undetected = [(p['x'], p['y']) for p in data_0['player_data'] if not p['is_detected']]
corners = data_0['image_corners_projection']
corner_x = np.array([corners['x_top_left'], corners['x_bottom_left'], corners['x_bottom_right'], corners['x_top_right'], corners['x_top_left']])
corner_y = np.array([corners['y_top_left'], corners['y_bottom_left'], corners['y_bottom_right'], corners['y_top_right'], corners['y_top_left']])
```

```python
ball = data_0['ball_data']
players_detected = [(p['x'], p['y']) for p in data_0['player_data'] if p['is_detected']]
players_undetected = [(p['x'], p['y']) for p in data_0['player_data'] if not p['is_detected']]

# Create pitch
pitch = Pitch(pitch_type='skillcorner', pitch_length=105, pitch_width=68)
fig, ax = pitch.draw(figsize=(10, 7))

# Plot players
pitch.scatter(*zip(*players_detected), ax=ax, c='green', label='Detected Players', s=100)
pitch.scatter(*zip(*players_undetected), ax=ax, c='red', label='Undetected Players', s=100)

# Plot ball
if ball['is_detected']:
    pitch.scatter(ball['x'], ball['y'], ax=ax, c='orange', label='Ball', s=200, edgecolors='black')

# Add legend
ax.legend(loc='upper left', fontsize=10)
print(corner_x)
print(corner_y)
# Show the plot
plt.show()
```

```python
def plot_frame(i):
    frame = data[i*100]
    ball = frame['ball_data']
    players_detected = [(p['x'], p['y']) for p in frame['player_data'] if p['is_detected']]
    players_undetected = [(p['x'], p['y']) for p in frame['player_data'] if not p['is_detected']]
    
    # Create pitch
    pitch = Pitch(pitch_type='skillcorner', pitch_length=105, pitch_width=68)
    fig, ax = pitch.draw(figsize=(6, 4))
    
    # Plot players
    pitch.scatter(*zip(*players_detected), ax=ax, c='green', label='Detected Players', s=10)
    pitch.scatter(*zip(*players_undetected), ax=ax, c='red', label='Undetected Players', s=10)
    
    # Plot ball
    if ball['is_detected']:
        pitch.scatter(ball['x'], ball['y'], ax=ax, c='orange', label='Ball', s=10, edgecolors='black')
    
    # Add legend
    ax.legend(loc='upper left', fontsize=10)
    # Show the plot
    plt.show()    
```

```python
for i in range(20):
    plot_frame(i)
```
